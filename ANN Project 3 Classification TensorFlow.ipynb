```python
#let us do our train test split 
X = df.drop("benign_0__mal_1", axis = 1).values 
y = df["benign_0__mal_1"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size
                            = 0.25, random_state = 101)
```

```python
X_train
```

```python
#Now let us scale the data using minmax scaler 
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
#X_train and X_test have been transformed to a scaled dataset
```

```python
X_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))
model.add(Dense(15, activation = "relu"))

#we are using Sigmoid at the output because it is a binary classifica
#tion problem, so the sigmoid function takes in real numbers and 
#output only a 1 or 0
#if dense was not 1, then we won't have been able to use sigmoid as 
#we are expecting one answer from 2 different option, in the case we 
#use Softmax which returns %, and we will take the max % as our ans.

model.add(Dense(1, activation = "sigmoid"))

model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
model.fit(x= X_train, y = y_train, epochs = 600, validation_data=(X_test, y_test) )
```

```python
model.history.history
```

```python
df_loss = pd.DataFrame(model.history.history)
```

```python
df_loss
```

```python
plt.figure(figsize= (12,6))
df_loss.plot()
```

```python
#it is obvious that the model is overfitting from the spike seen on 
#on the plot, so we will implore callback to stop the fitting once the
#validation loss set is going out of range
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))
model.add(Dense(15, activation = "relu"))

model.add(Dense(1, activation = "sigmoid"))

model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode = "min", 
                             verbose = 1, patience =25)
#what we are trying to do here is to stop the fitting when the 
#val_loss is going out of range, the monitor is monitoring the 
#validation loss, since that is out focus at the moment, while the 
#mode is min, since it is an error, we want to minimise as much as 
#possible but if the metric we were trying to measure was accuracy, 
#then we will look to returning the max value. patience is saying add
#25 more epochs to the operation after early stop has been detected
#because of the noise that could occur
```

```python
model.fit(x= X_train, y = y_train, epochs = 600, 
    validation_data=(X_test, y_test),callbacks=[earlyStopping]  )

#now we dont have to worry about the epoch we have selected since the 
#the callback will helps us to determine how epoch we actually need 
#to fit the data 
```

```python
model.history.history
```

```python
df_loss2 = pd.DataFrame(model.history.history)
```

```python
#another way to prevent overfitting is using the dropout function
#here we add a dropout to each of the layers by specifying how many
#percentage of the neurons we want to dropout
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))
model.add(Dropout(0.5)) #this means we are randomly dropping 50% of 
#the neurons we are taking 50% of the neurons and turning the off 
#during each batch of training, which in this case will be 1 epoch 
#at a time


model.add(Dense(15, activation = "relu"))
model.add(Dropout(0.5))

model.add(Dense(1, activation = "sigmoid"))

model.compile(optimizer = "adam", loss = "binary_crossentropy")
```

```python
model.fit(x= X_train, y = y_train, epochs = 600, 
    validation_data=(X_test, y_test),callbacks=[earlyStopping]  )
```

```python
df_loss_callback_dropout = pd.DataFrame(model.history.history)
```

```python
df_loss_callback_dropout
```

```python
df_loss_callback_dropout.plot()
```

```python
#now that we have trained our model very well, let us now use it to 
#predict
predictions = model.predict(X_test)
prediction=np.round(predictions).astype(int)
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, prediction))
print(confusion_matrix(y_test, prediction))
```
