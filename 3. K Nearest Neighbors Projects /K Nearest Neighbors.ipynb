```python
#this is another classification algorithm in machine learning 
#just like how we classified the linear and logistic regression into mean
#errors and classification report.
#for the KNN, we use the distance between the training dataset to predict
#here there are 4 steps in finding the kNN.
#1. store all the data
#2. calculate the distance of x to all the points in your data
#3. sort the points in your data by increasing the distance from x
#4. predict the majority labels on the points closest to K
#the advantages of KNN is that it is simple, training is trivial, more 
#data and classes could be added, and it invloves just few parameters such
#as the K and distance metrics 
#K is how many points are you going to look at that are close to your new  
#test points 
#the disadvantages are its predictions becomes less accurate as the amount
#of data increases.
#categorical features dont work well.
#when you given a dataset that the columns are classified, that you dont 
#know what each columns represent, it is called a take-home-task
```

```python
#when we are using KNN as classifier, we have to standardise the scale, 
#of all the features in the dataset aside the the target feature so that,
#there is not large from of each of the features from the target class
#now let us separate the data into X and y
X = df[["WTT","PTI","EQW","SBI","LQE","QWG","FDJ","PJF","HQE","NXJ"]]
```

```python
y = df["TARGET CLASS"]
```

```python
#now let us scale and transform the features 
from sklearn.preprocessing import StandardScaler
```

```python
model = StandardScaler()
```

```python
model.fit(X)
```

```python
transformed_feature = model.transform(X)
```

```python
transformed_feature
```

```python
#now let use the scaled and transformed dataset to recreate
#a new dataframe with the X.columns as the columns
df_transformed_feature = pd.DataFrame(transformed_feature,columns=X.columns)
```

```python
df_transformed_feature
```

```python
df_transformed_feature.head()
```

```python
y 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_transformed_feature,y,test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
#now we want to iterate the k values ie the n_neighbors values
#from 1 to 40
error_rate = []
for i in range(1,40):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    predictions_i = model.predict(X_test)
    error_rate.append(np.mean(predictions_i!=y_test))
    
print(error_rate)
```

```python
error_rate
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,40),error_rate,color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel = "K Value"
plt.ylabel = "Error_rate"
plt.title = "K Value VS Error rate"
```

```python
#now let us reassess the confusion matrix and classification report using
#different K that we received from the error rate
model = KNeighborsClassifier(34)
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
