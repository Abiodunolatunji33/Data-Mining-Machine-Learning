```python
#now we will look for a column that has few classification, so that it will
#be easy to classify the ages based on that column
```

```python
#then we can go ahead and boxplot the cols in respect to age to see the age
#distribution. let us look for the best column to use to classify the age.
#the higher the grouping of this column the better the age classification, 
#three groupings will be adequate, so we use class instead of age that
#has two groups
```

```python
Titanic_df
```

```python
sns.boxplot(x= "Pclass", y= "Age", data = Titanic_df)
```

```python
#from the above boxplot, we can classify the ages with the classes by
#let class1= 37, class2 = 29 and class3 = 24
def func(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
        
```

```python
Titanic_df["Age"] = Titanic_df[["Age","Pclass"]].apply(func,axis=1)
```

```python
X = Titanic_df.drop("Survived", axis = 1)
```

```python
y = Titanic_df["Survived"]
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(X)
```

```python
transformed_scaler = scaler.transform(X)
```

```python
transformed_scaler
```

```python
X_scaled = pd.DataFrame(transformed_scaler, columns = ["Pclass","Age","ibSp","Parch","Fare","male","Q","S"] )
```

```python
X_scaled
```

```python
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_scaled_train,y_train)
```

```python
Predictions = model.predict(X_test)
```

```python
Predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,Predictions))
print("\n")
print(classification_report(y_test,Predictions))
```
